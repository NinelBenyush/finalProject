```python
import pandas as pd
df = pd.read_csv("final_dataset.csv")
df.info()
```

```python
selected_columns = ['code_p', 'January_normalized', 'February_normalized','March_normalized','April_normalized','May_normalized','June_normalized','July_normalized','August_normalized','September_normalized','October_normalized','Novermber_normalized','December_normalized']
subset_df = df[selected_columns]
subset_df
```

```python
df2 = pd.melt(subset_df, id_vars=["code_p"], 
                  var_name="Month", value_name="Value")
```

```python
df2
```

```python
df2 = df2.sort_values(by=["code_p","Month"])
```

```python
df2
```

```python
month_map = {
    'January_normalized': 1, 'February_normalized': 2, 'March_normalized': 3, 'April_normalized': 4,
    'May_normalized': 5, 'June_normalized': 6, 'July_normalized': 7, 'August_normalized': 8,
    'September_normalized': 9, 'October_normalized': 10, 'Novermber_normalized': 11, 'December_normalized': 12
}

df2['month_numeric'] = df2['Month'].map(lambda x: month_map.get(x, None))

df2 = df2.sort_values(by=['code_p', 'month_numeric'])

df2 = df2.drop('month_numeric', axis=1)

```

```python
df2
```

```python
selected_columns = ['January', 'February', 'March', 'April', 'May', 'June', 'July_normalized', 'August', 'September', 'October', 'Novermber', 'December']
df_without_selected_columns = df.drop(columns=selected_columns)

df_without_selected_columns
```

```python
merged_df = pd.merge(df_without_selected_columns, df2, on='code_p')
merged_df.set_index('code_p', inplace=True)
```

```python
merged_df
```

```python
merged_df.to_csv("finally.csv",index=True)
```

```python
merged_df['Month'] = merged_df['Month'].str.replace('_normalized', '')
```

```python
merged_df
```

```python
new_column_names = [col.replace('_normalized', '') for col in merged_df.columns]

merged_df.columns = new_column_names
```

```python
merged_df
```

```python
merged_df.to_csv("finally2.csv")
```
