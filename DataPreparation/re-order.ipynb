```python
selected_columns = ['code_p', 'January', 'February','March','April','May','June','July','August','September','October','Novermber','December']
subset_df = df[selected_columns]
subset_df
```

```python
df2 = pd.melt(subset_df, id_vars=["code_p"], 
                  var_name="Month", value_name="Value")
df2
```

```python
df2 = df2.sort_values(by=["code_p","Month"])
df2
```

```python
month_map = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4,
    'May': 5, 'June': 6, 'July': 7, 'August': 8,
    'September': 9, 'October': 10, 'Novermber': 11, 'December': 12
}

df2['month_numeric'] = df2['Month'].map(lambda x: month_map.get(x, None))

df2 = df2.sort_values(by=['code_p', 'month_numeric'])

df2 = df2.drop('month_numeric', axis=1)
```

```python
selected_columns = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'Novermber', 'December']
df_without_selected_columns = df.drop(columns=selected_columns)

df_without_selected_columns
```

```python
merged_df = pd.merge(df_without_selected_columns, df2, on='code_p')
merged_df.set_index('code_p', inplace=True)
```
